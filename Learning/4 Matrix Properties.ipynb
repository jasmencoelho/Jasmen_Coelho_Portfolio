## Matrix Inversion
- approach for solving linear equations
- Matrix inverse of X is denoted as $X^{-1}$
- Satisfies: $X^{-1}$X = $I_{n}$

The regression formula can be represented as:
- y=Xw (w is the vector of weights a through m)


outcome vector = feature matrix * weights vector


In the equation y = Xw:
- we know the outcomes _y_, which could be house prices
- We know the features __X__, which are predictors like bedroom count
- Vector _w_ contains the unknowns, the model's learnable parameters

Assuming $X^{-1}$ exists, matrix inversion can solve for _w_:
- Xy = y
- $X^{-1}$Xw = $X^{-1}$y
- $I_{n}$w = $X^{-1}$y
- w = $X^{-1}$y

```python
X = np.array([[4, 2], [-5, -3]])
X
```

```python
Xinv = np.linalg.inv(X)
Xinv
```

```python
y = np.array([4,-7])
y
```

```python
w = np.dot(Xinv,y) # solution to weights
w
```

```python
np.dot(X,w) # checking by solving for y
```

Can only be calculated IF:
- Matrix isn't "singular"
- all columns of the matrix must be linearly independent


- Matrix must be square: $n_{row}$ = $n_{col}$
- vector span must equal the matrix range

## Diagonal Matrices 
- Nonzero elements along main diagonal, zeros everywhere else
- Identity matrix is an example
- If square, denotes as diag(x) where x is vector of main - diagonal elements
- Computationally efficient
- Multiplication: diag(x)y = np.dot(x,y)
- Inversion: $diag(x)^{-1}$ = $diag[1/x,...1/x]^{T}$ _Note: x can't include 0_
- Can be non-square: If height>width, simply add zeros to product, If width>height, remove elements from product


## Orthogonal Matrices

So, to sum it up, an orthogonal matrix is a special kind of matrix where the columns (or rows) are perpendicular to each other. It has properties like preserving vector lengths, easy invertibility, and representing geometric transformations. It's a powerful tool in mathematics and various practical fields.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Create a 3x3 orthogonal matrix
orthogonal_matrix = np.array([[0.8, 0.6, 0],
                              [-0.6, 0.8, 0],
                              [0, 0, 1]])

# Define the three axes vectors
x_axis = orthogonal_matrix[:, 0]
y_axis = orthogonal_matrix[:, 1]
z_axis = orthogonal_matrix[:, 2]

# Set up the figure and axes
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the vectors as arrows
origin = [0, 0, 0]
ax.quiver(origin[0], origin[1], origin[2], x_axis[0], x_axis[1], x_axis[2], color='r')
ax.quiver(origin[0], origin[1], origin[2], y_axis[0], y_axis[1], y_axis[2], color='g')
ax.quiver(origin[0], origin[1], origin[2], z_axis[0], z_axis[1], z_axis[2], color='b')

# Set the axis limits
ax.set_xlim([-1, 1])
ax.set_ylim([-1, 1])
ax.set_zlim([-1, 1])

# Set the labels
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Set the aspect ratio of the plot
ax.set_box_aspect([1, 1, 1])

# Show the plot
plt.show()
```
